```python
monthly_profit = df.groupby('YearMonth')['profit'].sum().reset_index()
monthly_profit['YearMonth'] = pd.to_datetime(monthly_profit['YearMonth'])
monthly_profit['RollingAvg'] = monthly_profit['profit'].rolling(window=3).mean()
monthly_profit
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
plt.plot(monthly_profit['YearMonth'], monthly_profit['profit'], label='Monthly Profit', marker='o')
plt.plot(monthly_profit['YearMonth'], monthly_profit['RollingAvg'], label='3-Month Rolling Avg', color='orange', linewidth=3)
plt.title("Monthly Profit Trend with Rolling Average", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Profit", fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
region_monthly_profit = df.groupby(['Region', 'YearMonth'])['profit'].sum().reset_index()
region_monthly_profit['YearMonth'] = pd.to_datetime(region_monthly_profit['YearMonth'])
region_monthly_profit['RollingAvg'] = (region_monthly_profit.groupby('Region')['profit'].transform(lambda x: x.rolling(window=3, min_periods=1).mean()))
plt.figure(figsize=(14, 7))
regions = region_monthly_profit['Region'].unique()
for region in regions:
    data = region_monthly_profit[region_monthly_profit['Region'] == region]
    plt.plot(data['YearMonth'], data['profit'], label=f'{region} (Actual)', linestyle='--', alpha=0.4)
    plt.plot(data['YearMonth'], data['RollingAvg'], label=f'{region} (3M Avg)', linewidth=2)
plt.title("Region-Wise Monthly Profit Trend with 3-Month Rolling Average", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Profit")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from pmdarima import auto_arima
```

```python
df = df.copy()
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['YearMonth'] = df['OrderDate'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('YearMonth')['Line Total'].sum().reset_index()
monthly_sales['YearMonth'] = pd.to_datetime(monthly_sales['YearMonth'])
monthly_sales['Sales (M)'] = monthly_sales['Line Total'] / 1000000
monthly_sales = monthly_sales.set_index('YearMonth').asfreq('MS')
auto_model = auto_arima(monthly_sales['Sales (M)'],
                        seasonal=True, m=12,        
                        stepwise=True, trace=True,
                        suppress_warnings=True)
n_periods = 12
forecast, conf_int = auto_model.predict(n_periods=n_periods, return_conf_int=True)
forecast_index = pd.date_range(start=monthly_sales.index[-1] + pd.offsets.MonthBegin(), periods=n_periods, freq='MS')
forecast_df = pd.DataFrame({'Forecast': forecast}, index=forecast_index)
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales['Sales (M)'], label='Historical Sales', marker='o')
plt.plot(forecast_df['Forecast'], label='Forecast (Auto SARIMA)', linestyle='--', color='orange')
plt.fill_between(forecast_index, conf_int[:, 0], conf_int[:, 1], color='orange', alpha=0.2, label='Confidence Interval')
plt.title("Auto-SARIMA: Forecast of Monthly Sales", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sales (Millions USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
